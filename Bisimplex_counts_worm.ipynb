```python
from robust_motifs.data import ResultManager
import seaborn as sns
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
r = ResultManager(Path("data/worm"))
```

```python
df = r.get_counts_dataframe("none")
```

```python
fig = plt.figure(figsize = [12,6])
ax = fig.add_subplot()
sns.lineplot(x = "dim", y = "count", data = df[df['motif'] == 'BS'], palette = 'Reds', hue = 'filename', ax= ax)
```

```python
fig = plt.figure(figsize = [12,6])
ax = fig.add_subplot()
sns.lineplot(x = "dim", y = "count", data = df[df['motif'] == 'ES'], palette = 'Reds', hue = 'filename', ax= ax)
```

```python
fig = plt.figure(figsize = [12,6])
ax = fig.add_subplot()
sns.lineplot(x = "dim", y = "count", data = df[df['motif'] == 'RBS+'], palette = 'Reds', hue = 'filename', ax= ax)
```

```python
fig = plt.figure(figsize = [12,6])
ax = fig.add_subplot()
sns.lineplot(x = "dim", y = "count", data = df[df['motif'] == 'RES+'], palette = 'Reds', hue = 'filename', ax= ax)
ax.set_yscale('log')
```
