```python
class hashable_array(np.ndarray):        
    def __hash__(self):
        return int.from_bytes(self.tobytes(), byteorder = "big")
    
    def __eq__(self, other):
        return hash(self) == hash(other)
```

```python
def hashable(matrix):
    return matrix
```

```python
matrix_table = {}
def recursive_n_simplices(matrix, hashing = True):
    if hashing:
        try:
            return matrix_table[hashable(matrix)]
        except KeyError:
            if matrix.shape[0] == 1:
                matrix_table[hashable(matrix)] = 1
                return 1
            res = np.sum([recursive_n_simplices(submatrix(matrix,y), hashing = True) for y in final_indices(matrix)])
            matrix_table[hashable(matrix)] = res
            return res
    else:
        if matrix.shape[0] == 1:
            return 1
        return np.sum([recursive_n_simplices(submatrix(matrix,y), hashing = False) for y in final_indices(matrix)])
```

```python
def final_indices(matrix):
    indices = []
    for i in range(matrix.shape[0]):
        j = i+1
        while j < matrix.shape[0]:
            if not matrix[i,j]:
                break
            j = j+1
        if j >= matrix.shape[1]:
            indices.append(i)
    return indices
```

```python
def submatrix(matrix, y):
    return np.delete(np.delete(matrix, y, axis = 0), y, axis = 1)
```

```python
def m1():
    n = 20
    test = hashable_array((n,n))
    test[:] = np.ones((n,n))
    return recursive_n_simplices(test, hashing = False)
```

```python
def m2():
    n = 20
    test = hashable_array((n,n))
    test[:] = np.ones((n,n))
    return recursive_n_simplices(test, hashing = True)
```

```python
timeit(m1, number = 1)
```

```python
timeit(m2, number = 1)
```
